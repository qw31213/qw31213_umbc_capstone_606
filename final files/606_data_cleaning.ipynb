```python
import pandas as pd

# Assuming your DataFrame is called df and the column is 'murmur_locations'

# Find the indices of rows where NaN values should be replaced with specific values
indices_100 = data[data['Murmur locations'] == 100].index
indices_300 = data[data['Murmur locations'] == 300].index

# Replace NaN values based on specific conditions
data.loc[indices_100, 'Murmur locations'] = 'MV+TV'
data.loc[indices_300, 'Murmur locations'] = 'AV+MV+PV+TV'

# For remaining NaN values, replace with 'MV+PV+TV'
data['Murmur locations'].fillna('MV+PV+TV', inplace=True)

print("DataFrame with NaN values replaced in the 'murmur_locations' column:")
print(data)
```

```python
import pandas as pd

# Assuming your DataFrame is called data and the column is 'Most audible location'

# Define the desired sample sizes
desired_samples = {
    'AV': 300,
    'MV': 200,
    'PV': 150,
    'TV': 114,
}

# Find the indices where 'Most audible location' column is NaN
nan_indices = data[data['Most audible location'].isna()].index

# Check the total number of available NaN rows
available_nans = len(nan_indices)

# Adjust sample sizes based on available NaN values
if sum(desired_samples.values()) > available_nans:
    ratio = available_nans / sum(desired_samples.values())
    desired_samples = {key: int(size * ratio) for key, size in desired_samples.items()}

# Initialize list to hold sampled indices
sampled_indices = []

# Sample indices for each replacement condition and remove sampled indices from available NaN indices
indices_av = data.loc[nan_indices].sample(n=desired_samples['AV'], replace=False).index
sampled_indices += indices_av.tolist()

remaining_indices = nan_indices.difference(sampled_indices)
indices_mv = data.loc[remaining_indices].sample(n=desired_samples['MV'], replace=False).index
sampled_indices += indices_mv.tolist()

remaining_indices = remaining_indices.difference(indices_mv)
indices_pv = data.loc[remaining_indices].sample(n=desired_samples['PV'], replace=False).index
sampled_indices += indices_pv.tolist()

remaining_indices = remaining_indices.difference(indices_pv)
indices_tv = data.loc[remaining_indices].sample(n=desired_samples['TV'], replace=False).index

# Replace NaN values based on specific conditions
data.loc[indices_av, 'Most audible location'] = 'AV'
data.loc[indices_mv, 'Most audible location'] = 'MV'
data.loc[indices_pv, 'Most audible location'] = 'PV'
data.loc[indices_tv, 'Most audible location'] = 'TV'

print("DataFrame with NaN values replaced in the 'Most audible location' column:")
print(data)
```

```python
import pandas as pd

# Assuming your DataFrame is called data and the column is 'Systolic murmur grading'

# Define the desired number of rows for each grading
desired_samples = {
    'I/VI': 324,
    'II/VI': 220,
    'III/VI': 220,
}

# Find the indices where 'Systolic murmur grading' column is NaN
nan_indices = data[data['Systolic murmur grading'].isna()].index

# Initialize list to hold the replacement values
replacement_values = []

# Generate replacement values based on the desired sample sizes for each grading
for grade, num_samples in desired_samples.items():
    replacement_values.extend([grade] * min(num_samples, len(nan_indices)))

# Replace NaN values with the generated replacement values
data['Systolic murmur grading'].fillna(pd.Series(replacement_values, index=nan_indices), inplace=True)

print("DataFrame with NaN values replaced in the 'Systolic murmur grading' column:")
print(data)
```

```python
import pandas as pd

# Assuming your DataFrame is called data and the column is 'Age'

# Define the desired number of rows for each age group
desired_samples = {
    'Child': 20,
    'Adolescent': 20,
    'Infant': 20,
    'Neonate': 15,
}

# Find the indices where 'Age' column is NaN
nan_indices = data[data['Age'].isna()].index

# Initialize list to hold sampled indices
sampled_indices = []

# Sample indices for each replacement condition and remove sampled indices from available NaN indices
for age_group, num_samples in desired_samples.items():
    # Sample indices for the current age group
    indices = data.loc[nan_indices].sample(n=num_samples, replace=False).index
    sampled_indices += indices.tolist()

# Replace NaN values based on specific conditions
data.loc[sampled_indices, 'Age'] = sampled_indices[:len(sampled_indices)]

print("DataFrame with NaN values replaced in the 'Age' column:")
print(data)
```

```python
import pandas as pd

# Assuming your DataFrame is called data and the column is 'Systolic murmur timing'

# Define the desired sample sizes
desired_samples = {
    'Early-systolic': 240,
    'Holosystolic': 240,
    'Mid-systolic': 90,
    'Late-systolic': 150,
}

# Find the indices where 'Systolic murmur timing' column is NaN
nan_indices = data[data['Systolic murmur timing'].isna()].index

# Check the total number of available NaN rows
available_nans = len(nan_indices)

# Adjust sample sizes based on available NaN values
if sum(desired_samples.values()) > available_nans:
    ratio = available_nans / sum(desired_samples.values())
    desired_samples = {key: int(size * ratio) for key, size in desired_samples.items()}

# Initialize list to hold sampled indices
sampled_indices = []

# Sample indices for each replacement condition and remove sampled indices from available NaN indices
for timing, num_samples in desired_samples.items():
    # Filter NaN values for the current timing
    timing_indices = nan_indices.intersection(data[data['Systolic murmur timing'].isna()].index)
    # Sample indices for the current timing
    if len(timing_indices) > 0:
        indices = data.loc[timing_indices].sample(n=min(num_samples, len(timing_indices)), replace=False).index
        sampled_indices += indices.tolist()

# Replace NaN values with specific timings based on sampled indices
data.loc[sampled_indices, 'Systolic murmur timing'] = sampled_indices[:len(sampled_indices)]

print("DataFrame with NaN values replaced in the 'Systolic murmur timing' column:")
print(data)
```
